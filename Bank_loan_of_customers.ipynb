```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Finance_1 = pd.read_csv('/content/Finance_1.csv')
Finance_1
```

```python
Finance_2 = pd.read_excel('/content/Finance_2.xlsx')
Finance_2
```

```python
df = pd.merge(Finance_1, Finance_2, on='id', how='inner')
df
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%y', errors='coerce')
df['issue_d']
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], errors='coerce')
df['last_pymnt_d']
```

```python
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.to_period('M')
df['last_pymnt_month'] = df['last_pymnt_d'].dt.to_period('M')

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

1.Year-wise Loan Amount Stats

```python
yearly_loan_stats = df.groupby('issue_year')['loan_amnt'].sum().reset_index()
plt.figure()
sns.barplot(x='issue_year', y='loan_amnt', data=yearly_loan_stats, palette='Blues_d')
plt.title("Year-wise Loan Amount")
plt.xlabel("Year")
plt.ylabel("Total Loan Amount")
ax = sns.barplot(x='issue_year', y='loan_amnt', data=yearly_loan_stats, palette='Blues_d')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2,
            height + 1000,
            f'{height:,.0f}',
            ha='center', va='bottom', fontsize=9, color='black')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2.Grade & Sub-grade wise revol_bal

```python
grade_revol = df.groupby(['grade', 'sub_grade'])['revol_bal'].sum().reset_index()
plt.figure(figsize=(14, 6))
sns.barplot(x='sub_grade', y='revol_bal', hue='grade', data=grade_revol)
plt.title("Grade and Sub-grade wise Revolving Balance")
plt.xlabel("Sub-grade")
plt.ylabel("Total Revolving Balance")
ax = sns.barplot(x='sub_grade', y='revol_bal', hue='grade', data=grade_revol)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width() / 2,
            height,
            f'{height:,.0f}',
            ha='center',
            va='bottom',
            fontsize=8,
            rotation=90
        )
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 3.Verified Status: Total Payment

```python
verified_payment = df.groupby('verification_status')['total_pymnt'].sum().reset_index()
plt.figure()
sns.barplot(x='verification_status', y='total_pymnt', data=verified_payment, palette='Set2')
plt.title("Total Payment by Verification Status")
plt.xlabel("Verification Status")
plt.ylabel("Total Payment")
ax = sns.barplot(x='verification_status', y='total_pymnt', data=verified_payment, palette='Set2')
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,
        height,
        f'{height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

4.State-wise & Month-wise Loan Status Using Stacked Plot Chart

```python
state_month_status = df.groupby(['addr_state', 'issue_month'])['loan_status'].value_counts().unstack().fillna(0).reset_index()
top_states = df['addr_state'].value_counts().head(5).index.tolist()
subset = df[df['addr_state'].isin(top_states)]
subset_status = subset.groupby(['issue_month', 'addr_state'])['loan_status'].value_counts().unstack().fillna(0)
subset_status = subset_status.reset_index().pivot(index='issue_month', columns='addr_state', values='Fully Paid').fillna(0)
subset_status.plot(kind='area', stacked=True, figsize=(14,6))
plt.title("Loan Status (Fully Paid) Over Time - Top 5 States")
plt.ylabel("Number of Loans")
plt.xlabel("Issue Month")
plt.tight_layout()
plt.show()
```

 5.Home Ownership vs Last Payment Date

```python
home_payment_stats = df.groupby(['home_ownership', 'last_pymnt_month'])['total_pymnt'].sum().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(x='last_pymnt_month', y='total_pymnt', hue='home_ownership', data=home_payment_stats)
plt.title("Home Ownership vs Last Payment Date")
plt.xlabel("Last Payment Month")
plt.ylabel("Total Payment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
